```python
%%javascript

//===helper functions
//variable to easily use canvas
var canvas = document.getElementById('canvas');

//function to print to output division
function print_div(s) { document.getElementById('out').innerHTML = s }

//remove #, parse, and return
function get_number(hex){  return parseInt(hex.replace('#',''), 16); }

//=== ini
let ctx;
let appWidth;
let appHeight;

//==='main'
function run_app(fetching=true) {
    
    sizeCanvas();
    
    print_div('getting...');

    if(fetching){
        //get from api
        NOOPBOT_FETCH({API: 'hexbot', count: 1000, width: appWidth, height: appHeight}, convert);
    }
    else{
        //get from sample file
        fetch('NOOP_SAMPLE.json')
            .then(response => response.json())
            .then(jsonResponse => convert(jsonResponse));
    }
}

//===functions

//use NOOP to setup canvas
function sizeCanvas() {

    appWidth = window.innerWidth;
    appHeight = window.innerHeight;
 
    ctx = NOOPBOT_SETUP_CANVAS( { canvas: canvas, bgColor:'#ffffff' } );
}

//read returned json (send to drawing function)
function convert(responseJson){
    let { colors } = responseJson;

    print_div('drawing...');
    
    colors.forEach(
        function(color,count) {
            //verify
            //divout.innerHTML += color.value + '(' + color.coordinates.x  + ',' + color.coordinates.y + '); ';  
            drawPoint(ctx, color, count);
    });
    
    print_div('fin.');
}

function drawPoint(ctx, color, count) {
   
    let pointSize = NOOPBOT_RANDOM(10,30);
    
    var color_as_number = get_number(color.value);
    var color_as_number_max = get_number('#ffffff');
    var color_x = color_as_number/color_as_number_max * appWidth;
    var color_y = color.coordinates.y;

    //draw path
    ctx.fillStyle = color.value;
    ctx.beginPath();
    ctx.arc(color_x , color_y, pointSize, 0, Math.PI * 2, true);
    ctx.fill();
}

//starts the app.
run_app(false);
```
